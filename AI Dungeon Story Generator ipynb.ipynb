```python
import streamlit as st
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load GPT-2 Model
@st.cache_resource
def load_model():
    tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
    model = GPT2LMHeadModel.from_pretrained("gpt2")
    return tokenizer, model

tokenizer, model = load_model()

st.title("🧙 AI Dungeon Story Generator")

st.write("Create interactive fantasy stories using Generative AI")

# Genre Selection
genre = st.selectbox(
    "Select Story Genre",
    ["Fantasy", "Mystery", "Adventure", "Sci-Fi", "Horror"]
)

# User Prompt
prompt = st.text_area("Enter the beginning of your story")

num_outputs = st.slider("Number of story continuations", 1, 5, 3)

max_length = st.slider("Story length", 50, 300, 150)

def generate_story(prompt_text):

    inputs = tokenizer.encode(prompt_text, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=max_length,
        num_return_sequences=num_outputs,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.9
    )

    stories = []

    for output in outputs:
        text = tokenizer.decode(output, skip_special_tokens=True)
        stories.append(text)

    return stories


if st.button("Generate Story"):

    if prompt == "":
        st.warning("Please enter a story prompt.")
    else:

        full_prompt = genre + " story: " + prompt

        results = generate_story(full_prompt)

        st.subheader("Generated Story Continuations")

        for i, story in enumerate(results):
            st.write(f"### Option {i+1}")
            st.write(story)

        # Save story option
        story_text = "\n\n".join(results)

        st.download_button(
            label="Download Story",
            data=story_text,
            file_name="ai_dungeon_story.txt",
            mime="text/plain"
        )
```
